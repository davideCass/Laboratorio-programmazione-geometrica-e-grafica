```python
## leggo i vertici dal file .lines e li trascrivo in un dizionario
from collections import defaultdict

with open("pianta_villa_cornaro.lines") as f:
    V = defaultdict(list)
    i = 0
    j = 0    
    for line in f:
        for e in line.split(","):
            V[i].append(round(float(e),7))       ## scrivo la coordinata e la tronco alle prima 7 cifre dopo la virgola
            j = j+1
            if(j%2==0):    
                i = i+1

## creo un dizionario di segmenti dai vertici V nella forma [[x1, y1, x2, y2]]

EV = defaultdict(list)
i = 0
k = 0
while(V[k+1]!=[]):
    EV[i].append([[V[k][0], V[k][1]], [V[k+1][0], V[k+1][1]]])
    k = k+2    
    i = i+1
print(EV)         
```
